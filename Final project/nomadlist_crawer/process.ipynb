```python
#Country Year
import numpy as np
import csv

countries = {}
def time(date_month_year):
    tmp = date_month_year.split()
    
    if(len(tmp)==2):
        tmp.append('2018')
    else:
        y = int(tmp[2][1:])
        if(y>18):
            tmp[2] = '19'+tmp[2][1:]
        else:
            tmp[2] = '20'+tmp[2][1:]
    return {"date":tmp[0],"month":tmp[1],"year":tmp[2]}
    
with open("UserTrips.csv") as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            uid = row[0].strip()
            if(row[1]!=''):
                start = time(row[1])
            else:
                continue
            city = row[3]
            country = row[4].strip()
            
            if(country in countries):
                if(start["year"] in countries[country]):
                    countries[country][start["year"]].append(uid)
                else:
                    countries[country][start["year"]] = [uid]
            else:
                countries[country] = {
                    start["year"]: [uid]
                }
    
    f = open('CountryYearNum.csv', 'a')
    writer = csv.writer(f)
    writer.writerow(['country', 'year', 'num'])
    for c in countries.keys():
        for y in countries[c].keys():
            num = len(set(countries[c][y]))
            writer.writerow([c,y,num])
```

```python
# Country trend over 2000-2018 (years)
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
countries = {}
with open("CountryYearNum.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')
    for i, row in enumerate(data):
        if(i>0):
            country = row[0]
            year = row[1]
            num = row[2]
            if(country in countries):
                countries[country][int(year)-2000]=int(num)
            else:
                countries[country] = np.zeros([19])

X = np.array([2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])
fig = plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
ax = plt.subplot(111)
formatter = ticker.StrMethodFormatter("{x:.0f}")
ax.xaxis.set_major_formatter(formatter)

plt.bar(X-0.3, countries['Taiwan'], width=0.3,facecolor='#9999ff', edgecolor='white', label='Taiwan')
plt.bar(X, countries['Thailand'], width=0.3, facecolor='#ff9999', edgecolor='white', label='Thailand')
plt.bar(X+0.3, countries['United States'], width=0.3, facecolor='#99ff99', edgecolor='white', label='United States')
plt.legend(loc='best')
plt.title("# of Nomads Visited Taiwan, Thailand, USA in 2000-2018")
plt.xlabel("Year")
plt.ylabel("# of Nomads")
plt.savefig('Thailand00-18.png')
```

```python
# Countries 6 year a slot 2000-2018
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import math
countries = {}
with open("data/CountryYearNum.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')
    for i, row in enumerate(data):
        if(i>0):
            country = row[0]
            year = int(row[1])
            num = int(row[2])
            if(year>=2000 and year<2018):
                slot = math.floor((year-2012)/6)
                if(country in countries):
                    
                    countries[country][slot]+=num
                    
                else:
                    countries[country]=np.zeros([3], dtype=int)
                    countries[country][slot] = num
print(countries)
f = open('City-6-YearNum.csv', 'w')
writer = csv.writer(f)
writer.writerow(['country', '2000-2005','2006-2011','2012-2017'])
for c in countries.keys():
    writer.writerow([c,countries[c][0],countries[c][1],countries[c][2]])
```

```python
#Find country pattern

countries = set()
with open("Data/data.csv", 'r') as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            country = row[7]
            countries.add(country)
f.close()
country_dic = {}
for i,c in enumerate(countries):
    country_dic[c] = str(i)
print(country_dic)

people = {}
output_data = []
with open("Data/UserTrips.csv") as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            uid = row[0].strip()
            country = row[4].strip()
            if(country in countries):
                if(uid in people):
                    people[uid].append(country_dic[country])
                else:
                    people[uid] = [country_dic[country]]
with open("UserCountryPattern.txt", 'w') as f:
    for p in people.keys():
        value = set(people[p])
        s = ",".join(value )
        f.write(s+'\n')
    
```
