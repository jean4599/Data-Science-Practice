```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
import csv

userid = []
trips_count = []
countries_count = []
cities_count = []

with open('UserProfile.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            userid.append(row[0])
            trips_count.append(int(row[1]))
            countries_count.append(int(row[2]))
            cities_count.append(int(row[3]))
print(max(trips_count))
plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.hist(trips_count)
plt.title("Trip Count")
plt.xlabel("# of Trip")
plt.ylabel("Frequency")
plt.savefig('tripcount.png')
plt.show()

plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.hist(countries_count)
plt.title("Country Count")
plt.xlabel("# of Country")
plt.ylabel("Frequency")
plt.savefig('contrycount.png')
plt.show()

plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.hist(cities_count)
plt.title("City Count")
plt.xlabel("# of City")
plt.ylabel("Frequency")
plt.savefig('citycount.png')
plt.show()
```

```python
#Country month
import numpy as np
import csv
import matplotlib.pyplot as plt
countries = {}
def time(date_month_year):
    tmp = date_month_year.split()
    if(tmp[1]=='Jan'):
        tmp[1] = 1
    elif(tmp[1]=='Feb'):
        tmp[1] = 2
    elif(tmp[1]=='Mar'):
        tmp[1] = 3
    elif(tmp[1]=='Apr'):
        tmp[1] = 4
    elif(tmp[1]=='May'):
        tmp[1] = 5
    elif(tmp[1]=='Jun'):
        tmp[1] = 6
    elif(tmp[1]=='Jul'):
        tmp[1] = 7
    elif(tmp[1]=='Aug'):
        tmp[1] = 8
    elif(tmp[1]=='Sep'):
        tmp[1] = 9
    elif(tmp[1]=='Oct'):
        tmp[1] = 10
    elif(tmp[1]=='Nov'):
        tmp[1] = 11
    elif(tmp[1]=='Dec'):
        tmp[1] = 12
    return {"date":tmp[0],"month":tmp[1]}
    
with open("UserTrips.csv") as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            uid = row[0].strip()
            start = time(row[1])
            end = time(row[1])
            city = row[3]
            country = row[4].strip()
            
            if(country in countries):
                if(start["month"] in countries[country]):
                    countries[country][start["month"]].append(uid)
                else:
                    countries[country][start["month"]] = [uid]
            else:
                countries[country] = {
                    start["month"]: [uid]
                }
    
#     f = open('CountryMonthNum.csv', 'a')
#     writer = csv.writer(f)
#     writer.writerow(['country', 'year', 'num'])
#     for c in countries.keys():
#         for m in countries[c].keys():
#             num = len(set(countries[c][m]))
#             writer.writerow([c,m,num])

X = range(12)
Y1 = np.zeros([12])
Y2 = np.zeros([12])

plt.title('Number of Visited Nomads Across 12 Months')
plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.xlabel('Mounth')
plt.ylabel('# of Nomads')
for m in countries['Taiwan'].keys():
    Y1[int(m)-1] = len(countries['Taiwan'][m])
for m in countries['Thailand'].keys():
    Y2[int(m)-1] = len(countries['Thailand'][m])

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white', label='Taiwan')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white', label='Tailand')
plt.legend(loc='best')
plt.savefig('Taiwan-Tailand.png')
plt.show()
```

```python
#Country Year
import numpy as np
import csv

countries = {}
def time(date_month_year):
    tmp = date_month_year.split()
    
    if(len(tmp)==2):
        tmp.append('2018')
    else:
        y = int(tmp[2][1:])
        if(y>18):
            tmp[2] = '19'+tmp[2][1:]
        else:
            tmp[2] = '20'+tmp[2][1:]
    return {"date":tmp[0],"month":tmp[1],"year":tmp[2]}
    
with open("UserTrips.csv") as f:
    reader = csv.reader(f, delimiter=',')
    for i, row in enumerate(reader):
        if(i>0):
            uid = row[0].strip()
            if(row[1]!=''):
                start = time(row[1])
            else:
                continue
            city = row[3]
            country = row[4].strip()
            
            if(country in countries):
                if(start["year"] in countries[country]):
                    countries[country][start["year"]].append(uid)
                else:
                    countries[country][start["year"]] = [uid]
            else:
                countries[country] = {
                    start["year"]: [uid]
                }
    
    f = open('CountryYearNum.csv', 'a')
    writer = csv.writer(f)
    writer.writerow(['country', 'year', 'num'])
    for c in countries.keys():
        for y in countries[c].keys():
            num = len(set(countries[c][y]))
            writer.writerow([c,y,num])
```

```python
# Country trend over 2000-2018 (years)
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
countries = {}
with open("CountryYearNum.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')
    for i, row in enumerate(data):
        if(i>0):
            country = row[0]
            year = row[1]
            num = row[2]
            if(country in countries):
                countries[country][int(year)-2000]=int(num)
            else:
                countries[country] = np.zeros([19])

X = np.array([2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])
fig = plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
ax = plt.subplot(111)
formatter = ticker.StrMethodFormatter("{x:.0f}")
ax.xaxis.set_major_formatter(formatter)

plt.bar(X-0.3, countries['Taiwan'], width=0.3,facecolor='#9999ff', edgecolor='white', label='Taiwan')
plt.bar(X, countries['Thailand'], width=0.3, facecolor='#ff9999', edgecolor='white', label='Thailand')
plt.bar(X+0.3, countries['United States'], width=0.3, facecolor='#99ff99', edgecolor='white', label='United States')
plt.legend(loc='best')
plt.title("# of Nomads Visited Taiwan, Thailand, USA in 2000-2018")
plt.xlabel("Year")
plt.ylabel("# of Nomads")
plt.savefig('Thailand00-18.png')
```
